```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import KBinsDiscretizer
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gabetrix/public-datasets/main/Iris.csv')
df.head()
```

```python
df.shape
```

```python
te = TransactionEncoder()
te_ary = te.fit(df).transform(df)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
apriori(df, min_support=0.1)
```

```python
iris = load_iris()
X = iris.data
columns = iris.feature_names
df = pd.DataFrame(X, columns=columns)
discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
X_discretized = discretizer.fit_transform(X)
df_discretized = pd.DataFrame(X_discretized, columns=columns)
print(df_discretized.head())
df_discretized.describe()
```

```python
transactions = []
for index, row in df_discretized.iterrows():
  transaction = []
  for column in df_discretized.columns:
    transaction.append((column, row[column]))
  transactions.append(transaction)
```

```python
df_bin = pd.get_dummies(df, columns=df_discretized.columns)
r_apriori = apriori(df, min_support=0.1, use_colnames=True)
```

```python
df_bin = pd.get_dummies(df_discretized, columns=df_discretized.columns)
frequent_itemsets = apriori(df_bin, min_support=0.3, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print(rules)
```
